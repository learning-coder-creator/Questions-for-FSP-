```python
import numpy as np

data = [10, 20, 30, 40, 50]

# Calculate Mean
mean = np.mean(data)
print("Mean =", mean)

# Calculate Standard Deviation
std_dev = np.std(data)
print("Standard Deviation =", std_dev)

#  Compute Z-scores
z_scores = []

for x in data:
    z = (x - mean) / std_dev
    z_scores.append(z)

# Display transformed data
print("Original Data:", data)

print("Z-score Normalized Data:")

for i in range(len(data)):
    print(data[i], "->", round(z_scores[i], 2))
```
