```python
from sklearn.preprocessing import StandardScaler
import numpy as np

dataset = np.array([
    [10, 1000],
    [20, 2000],
    [30, 3000],
    [40, 4000]
])

print("Original Data:")
print(dataset)

scaler = StandardScaler()
scaled_dataset = scaler.fit_transform(dataset)

print("\nStandardized Data:")
print(scaled_dataset)
```
