```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import numpy as np

dataset = np.array([
    [5, 50],
    [10, 100],
    [15, 150],
    [20, 200],
    [25, 250]
])

min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

min_max_output = min_max_scaler.fit_transform(dataset)
standard_output = standard_scaler.fit_transform(dataset)

# Display the comparisons
print("Original Dataset:")
print(dataset)

print("\nMin-Max Scaled Output (Values between 0 and 1):")
print(min_max_output)

print("\nStandardized Output (Mean = 0, Standard Deviation = 1):")
print(standard_output)
```
