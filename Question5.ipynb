```python
import numpy as np

def detect_outliers_zscore(data, threshold=3.0):
    
    # Calculate the mean and standard deviation of the dataset
    mean = np.mean(data)
    std_dev = np.std(data)
    
    outliers = []
    
    for x in data:
        z_score = (x - mean) / std_dev
        if np.abs(z_score) > threshold:
            outliers.append(x)
            
    return outliers


dataset = [10, 12, 11, 15, 11, 14, 13, 12, 10, 11, 99, 12, 14]

detected_outliers = detect_outliers_zscore(dataset)

print(f"Original Dataset:  {dataset}")
print(f"Detected Outliers: {detected_outliers}")
```
