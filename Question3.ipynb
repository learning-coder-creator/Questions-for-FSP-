```python
def min_max_normalize(data):
    x_min = min(data)
    x_max = max(data)
    
    if x_max == x_min:
        return [0.0 for _ in data]
    
    normalized_data = [(x - x_min) / (x_max - x_min) for x in data]
    
    return normalized_data

values = [12, 25, 40, 55, 70]
normalized_values = min_max_normalize(values)

print(f"Original values:   {values}")
print(f"Normalized values: {normalized_values}")
```
