```python

import pandas as pd
import numpy as np


data = {
    "Name": ["Amit", "Riya", "Sohan", "Neha", "Karan"],
    "Marks": [85, np.nan, 78, 90, np.nan],
    "Age": [20, 21, np.nan, 22, 23]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Detect missing values
print("\nMissing Values in DataFrame:")
print(df.isnull())

# Replace missing values with mean
df_mean = df.copy()
df_mean["Marks"] = df_mean["Marks"].fillna(df_mean["Marks"].mean())
df_mean["Age"] = df_mean["Age"].fillna(df_mean["Age"].mean())

print("\nDataFrame after replacing missing values with Mean:")
print(df_mean)

# Replace missing values with median
df_median = df.copy()
df_median["Marks"] = df_median["Marks"].fillna(df_median["Marks"].median())
df_median["Age"] = df_median["Age"].fillna(df_median["Age"].median())


print("\nDataFrame after replacing missing values with Median:")
print(df_median)

# Drop rows containing missing values
df_drop = df.dropna()

print("\nDataFrame after dropping rows with missing values:")
print(df_drop)
```
